```python
# OCR (Optical Character Recognition) 
from pdf2image import convert_from_path
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
# Convert into images
pages = convert_from_path(
                        pdf_path="data/HSC26-Bangla1st-Paper.pdf", 
                        dpi=300,
                        fmt = "png",
                        thread_count=4,
                        grayscale=False,
                        use_cropbox=False,
                        strict=False
)
```

```python
len(pages)
```

```python
# convert images into raw text
import pdfplumber


raw_text = ""

table_number = [2, 41]

for i,page in  enumerate(pages):
    if i in table_number:
        with pdfplumber.open("data\HSC26-Bangla1st-Paper.pdf") as pdf:
            table_page = pdf.pages[i - 1]  # select page
            tables = table_page.extract_tables()

            for table in tables:
                raw_text += "উপরে দেওয়া প্রশ্নগুলোর সঠিক উত্তরগুলো প্রশ্ন নম্বর অনুযায়ী দেওয়া আছে, যেমন ‘১’, ‘ক’ (SL, ANS)" + '\n'
                for row in table:
                    raw_text += str(row) + '\n'
        
    text = pytesseract.image_to_string(page, lang ="ben+eng", config='--psm 6 --oem 1')
    raw_text += text
```

## Step 2


```python
## create document object
from langchain.docstore.document import Document

cleaned_doc = Document(page_content=cleaned_text, metadata={"source": "HSC26-Bangla1st-Paper.pdf"})
```

```python
cleaned_doc
```

```python
## text splitter
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 300,
    chunk_overlap = 35,
    separators=["\n\n", "\n", " ", ""]
)
```

```python
chunks = text_splitter.split_documents([cleaned_doc])
```

```python
chunks
```

## step 2

```python
from langchain_huggingface import HuggingFaceEmbeddings, HuggingFaceEndpoint, ChatHuggingFace
from langchain.vectorstores import FAISS
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name = "l3cube-pune/bengali-sentence-similarity-sbert"
)
```

```python
vectore_store = FAISS.from_documents(
    documents= chunks,
    embedding=embeddings
)
```

```python
vectore_store.save_local(
    folder_path="vectore_store"
    
)
```

```python
load_vector = FAISS.load_local(
    folder_path="vectore_store",
)
```

```python
query = " বিয়ের সময় কল্যাণীর প্রকৃত বয়স কত ছিল?"
```

```python
query_vector = embeddings.embed_query(query)

# 4. Search with score
results = vectore_store.similarity_search_with_score_by_vector(
    query_vector, k=10
)

results
```

```python
retriever = vectore_store.as_retriever(
    search_type="similarity",
    search_kwargs={
        "k": 10,                      # Top 5 most similar chunks
        "score_threshold": 0.5,      # Optional: Filter out irrelevant matches (lower is stricter)
        "include_metadata": True,    # Return page/source info with chunks
        "fetch_k": 20                # Search from top-20 candidates before narrowing to top-5
    }
)
```

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory, InMemoryChatMessageHistory
from langchain.schema.runnable import RunnableLambda
```

```python
store = {}
```

```python
def get_by_session_id(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]
```

```python
system_prompt = (
    "You are a professional AI assistant specialized in the HSC 1st Paper Bangla textbook, "
    "which contains stories, comprehension questions (MCQs), and detailed model answers. "
    "Your primary role is to provide accurate, concise answers strictly based on the provided context below, "
    "especially focusing on story details, character information, and specific question-answer pairs. "
    "Provide precise and unambiguous answers only. False or vague information may lead to confusion, so always be specific. "
    "If the answer is not clearly present in the context, respond with: 'The information is not available in the provided context.' "
    "Never provide any false, assumed, or fabricated information. If unsure, say the answer is not in the context. "
    "When Bengali literary terms or story elements appear in the context, briefly explain them simply and clearly, with Bengali examples if helpful. "
    "Keep all responses short (maximum three sentences) and precise to the question asked. "
    "You may understand and respond to user messages in mixed language, but respond in the same language script as the user: "
    "if the user asks in pure Bangla script (e.g., 'অনুপমের ভাষায় সুপুরুষ কাকে বলা হয়েছে?'), answer in Bangla script; "
    "if the user asks in English, answer in English. "
    "Always include Bengali names or terms as they appear in the text. "
    "If the user provides MCQ answers as a raw list format like ['SL', 'Ans', ...] such as ['১', 'খ', '২', 'গ'], "
    "then reformat them into clear Bangla sentences like: '১ নম্বর প্রশ্নের সঠিক উত্তর: খ)', one per line. "
    "But if the user asks an MCQ-style question in natural language (e.g., ‘অনুপমের বয়স কত?’), simply provide the factual answer directly without using 'SL' or option letters."
    "\n\n"
    "{context}"
)
```

```python
# prompt
prompt = ChatPromptTemplate.from_messages([
    ('system' , system_prompt),
    MessagesPlaceholder(variable_name="chat_history"),
    ('human', "{input}")
])

# create chain
qa_chain = create_stuff_documents_chain(llm = model, prompt=prompt)
rag_chain = create_retrieval_chain(retriever, qa_chain)
wrapped_chain = rag_chain | RunnableLambda(lambda output: output["answer"])
```

```python
chain_with_history = RunnableWithMessageHistory(
    wrapped_chain,
    get_by_session_id,
    input_messages_key="input",
    history_messages_key="chat_history"
)
```

```python
result = chain_with_history.invoke(
    {"input" : "অনুপমের ভাষায় সুপুরুষ কাকে বলা হয়েছে?"}, # q here
    config = {
        "configurable": {"session_id": "yoo"}
    }
)
```
